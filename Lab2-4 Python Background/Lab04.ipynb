```python
titles = ['square', 'square root', 'exp', 'log']
figure, a = plt.subplots(2,2)
x = arange(1,10)
graphs = [x*x, sqrt(x), exp(x), log10(x)]
k = 0
for i in range(2):
    for j in range(2):
        a[i][j].plot(x, graphs[k])
        a[i][j].set_title(titles[k], fontdict = font2)
        k += 1
plt.suptitle('Math basic graphs', fontdict=font1)
```

```python
a1 = plt.subplot2grid((3,3), (0,0), colspan = 2)
a2 = plt.subplot2grid((3,3), (0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3), (1,0), colspan = 2, rowspan = 2)

x = arange(1, 10)
a1.plot(x, x*x)
a2.plot(x, log10(x))
a3.plot(x, exp(x))

plt.tight_layout()
```

```python
distribution = array([35, 25,15, 25])
labels = array(['Won', 'Lost', 'Rain', 'NA'])
explode = array([0.05, 0.05, 0.05, 0.05])
colors = array(['Green', 'Red',"Pink", "Yellow"])
plt.pie(distribution,
       labels = labels,
       startangle = 90,
       explode = explode,
       shadow = True,
       colors = colors,
       autopct = '%1.f%%' 
       )
plt.legend(title = 'Matches Status')
```
